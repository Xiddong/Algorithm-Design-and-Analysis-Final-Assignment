## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

### idea:

我先固定买入日，循环卖出日。

枚举第一日买入第二日卖出、第一日买入第三日卖出、第一日买入第四日卖出……第二日买入第三日卖出、第二日买入第四日卖出以此类推。

每次更新收益的最大值。

外层循环枚举买入日，内层循环枚举卖出日，所以大概会比较 $n^2$ 次，时间复杂度是 $O(n^2)$。

````python
def max(profits):
    n = len(profits)
    result = profits[0]
    buy = 0
    sell = 0

    for i in range(n):
        cur = 0
        for j in range(i, n):
            cur += profits[j]
            if cur > result:
                result = cur
                buy = i
                sell = j

    return result, buy + 1, sell + 1

profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
result = max(profits)
print('profits=[3,2,1,-7,5,2,-1,3,-1], 第' + str(result[1]) + '天买入, 第' + str(result[2]) + '天卖出, 收益最大：' + str(result[0]))
````

## 问题 4

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

### idea:

这题先用二分找一个局部最小值。

我每次看中间的数，如果它比左右两边都小，就直接返回。

如果左边更小，就往左边继续找；否则就往右边继续找。

例子里面 3 和 1 都是局部最小值，所以我又写了一个简单循环，把所有局部最小值也找出来。

二分的时候每次都会少看一半，所以找一个局部最小值是 $O(\log n)$；找所有局部最小值要把数组扫一遍，所以是 $O(n)$。

````python
def find(A):
    left = 1
    right = len(A) - 2

    while left <= right:
        mid = (left + right) // 2

        if A[mid] < A[mid - 1] and A[mid] < A[mid + 1]:
            return A[mid]
        elif A[mid - 1] < A[mid]:
            right = mid - 1
        else:
            left = mid + 1

    return None


def findall(A):
    result = []
    for i in range(1, len(A) - 1):
        if A[i] < A[i - 1] and A[i] < A[i + 1]:
            result.append(A[i])
    return result


A = [9, 3, 7, 2, 1, 4, 5]
result = findall(A)
print('A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 ' + str(result[0]) + ', ' + str(result[1]))
````

## 问题 5

给定包含n个不同数字的一组数, 寻找一种基于比较的算法在这组数中找到k个最小的数字, 并按顺序输出它们。

1. 将n个数先进行排序, 然后按顺序输出最小的k个数。要求：选择合适的排序算法实现上述操作, 计算算法时间复杂度, 并使用python编程实现。
2. 建立一个包含这n个数的堆（heap）, 并且调用 k 次Extract-min 按顺序输出最小的k个数。使用往空堆中不断插入元素的方法建立堆, 分析这种方法建堆的时间复杂度, 并使用python编程实现
3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

### idea:

第一种方法是直接排序。

我先把数组排好序，然后输出前 $k$ 个数。归并排序每次把数组分成两半，再合并回来，所以时间复杂度是 $O(n\log n)$。

第二种方法是建一个小根堆。

先把所有数一个一个插入 heap 里，再调用 $k$ 次 `extract_min`。插入一个数最多调整一条堆上的路径，所以建堆大概是 $O(n\log n)$，后面弹出 $k$ 次是 $O(k\log n)$。

````python
def merge(arr):
    if len(arr) <= 1:
        return arr[:]

    mid = len(arr) // 2
    left = merge(arr[:mid])
    right = merge(arr[mid:])

    result = []
    i = 0
    j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result


def sort(arr, k):
    return merge(arr)[:k]


class heap:
    def __init__(self):
        self.arr = []

    def insert(self, x):
        self.arr.append(x)
        i = len(self.arr) - 1
        while i > 0:
            father = (i - 1) // 2
            if self.arr[father] <= self.arr[i]:
                break
            self.arr[father], self.arr[i] = self.arr[i], self.arr[father]
            i = father

    def extract_min(self):
        if not self.arr:
            return None
        result = self.arr[0]
        last = self.arr.pop()
        if self.arr:
            self.arr[0] = last
            self.down(0)
        return result

    def down(self, i):
        n = len(self.arr)
        while True:
            small = i
            left = 2 * i + 1
            right = 2 * i + 2

            if left < n and self.arr[left] < self.arr[small]:
                small = left
            if right < n and self.arr[right] < self.arr[small]:
                small = right

            if small == i:
                break
            self.arr[i], self.arr[small] = self.arr[small], self.arr[i]
            i = small


def use_heap(arr, k):
    h = heap()
    for x in arr:
        h.insert(x)

    result = []
    for _ in range(k):
        result.append(h.extract_min())
    return result

arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
result = use_heap(arr, k)
print('数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：' + str(result[0]) + ', ' + str(result[1]) + ', ' + str(result[2]))
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

### idea:

我用一个栈保存还没有被撞掉的星球。

如果栈顶是正数，当前星球是负数，就说明它们可能会撞。

谁的绝对值小谁消失，如果一样大，题目说负数那个消失。

每个星球最多进栈一次，也最多被弹出一次，所以总的操作次数和星球数量差不多，时间复杂度是 $O(n)$。

````python
def hit(A):
    stack = []

    for x in A:
        survive = True
        while survive and stack and stack[-1] > 0 and x < 0:
            if stack[-1] < -x:
                stack.pop()
            else:
                survive = False
        if survive:
            stack.append(x)

    return stack

A1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
A2 = [23, -8, 9, -3, -7, 9, -23, 22]
print('输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：' + str(hit(A1)))
print('输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：' + str(hit(A2)))
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

### idea:

这题是 0-1 背包。

每块大理石只有两种情况：装或者不装。

我用 `dp[i][j]` 表示前 `i` 块石头、容量为 `j` 时的最大价值。

每次比较装第 `i` 块和不装第 `i` 块哪个更大。

因为要把每个石头和每个容量都算一遍，所以如果有 $n$ 块石头、容量是 $W$，时间复杂度是 $O(nW)$。

````python
def bag(stone, size):
    n = len(stone)
    dp = [[0] * (size + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        weight, value = stone[i - 1]
        for j in range(size + 1):
            dp[i][j] = dp[i - 1][j]
            if j >= weight and dp[i - 1][j - weight] + value > dp[i][j]:
                dp[i][j] = dp[i - 1][j - weight] + value

    return dp[n][size]

stone = [(5, 10), (4, 40), (6, 30), (3, 50)]
size = 9
print('最大价值:', bag(stone, size))
````

### idea:

二叉搜索树中序遍历应该是从小到大。

现在有两个点被换了，所以中序遍历的时候会出现顺序不对的地方。

我先把中序遍历的节点存到数组里，再找逆序的位置。

找到两个出错的节点以后，交换它们的值就行，树的形状不改。

中序遍历会访问每个节点一次，后面找错误节点也只是扫一遍数组，所以时间复杂度是 $O(n)$。

````python
class Node:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def inittree(arr):
    if not arr or arr[0] is None:
        return None

    node = [None if x is None else Node(x) for x in arr]
    child = 1
    for i in range(len(node)):
        if node[i] is not None:
            if child < len(node):
                node[i].left = node[child]
                child += 1
            if child < len(node):
                node[i].right = node[child]
                child += 1
    return node[0]


def cout(root):
    if root is None:
        return []

    result = []
    queue = [root]
    while queue:
        cur = queue.pop(0)
        if cur is None:
            result.append(None)
        else:
            result.append(cur.val)
            queue.append(cur.left)
            queue.append(cur.right)

    while result and result[-1] is None:
        result.pop()
    return result


def fix(root):
    arr = []

    def mid(root):
        if root is None:
            return
        mid(root.left)
        arr.append(root)
        mid(root.right)

    mid(root)

    first = None
    second = None
    for i in range(len(arr) - 1):
        if arr[i].val > arr[i + 1].val:
            if first is None:
                first = arr[i]
            second = arr[i + 1]

    if first is not None and second is not None:
        first.val, second.val = second.val, first.val

    return root

r1 = inittree([1, 3, None, None, 2])
fix(r1)
print('输入：root = [1,3,null,null,2]， 输出为：' + str(cout(r1)).replace('None', 'null').replace(' ', ''))

r2 = inittree([3, 1, 4, None, None, 2])
fix(r2)
print('输入： root=[3,1,4,null,null,2]， 输出为：' + str(cout(r2)).replace('None', 'null').replace(' ', ''))
````

给你一个二叉树的根节点 root ，树中每个节点都存放有一个 0 到 9 之间的数字。每条从根节点到叶节点的路径都代表一个数字：

设计一个算法计算从根节点到叶节点生成的所有数字之和 ，分析算法设计思路，计算时间复杂度，并基于python编程实现

叶节点:是指没有子节点的节点。

e.g.:
---
1.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-1.jpg"> </div>

 输入：root = [1,2,3], 输出：25

解释：从根到叶子节点路径 1->2 代表数字 12，1->3 代表数字 13，因此和为 12 + 13 = 25 。

2.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-2.jpg"> </div>

 输入：root = [4,9,0,5,1], 输出：1026

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。

### idea:

我用递归从根节点往下走。

每走到一个节点，就把前面的数字乘以 10，再加上当前节点的值。

如果走到叶子节点，就说明得到了一条完整的数字，把它返回。

最后把所有叶子节点返回的数字加起来。

递归时每个节点只会被访问一次，所以时间复杂度是 $O(n)$。

````python
class Node:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def inittree(arr):
    if not arr or arr[0] is None:
        return None

    node = [None if x is None else Node(x) for x in arr]
    child = 1
    for i in range(len(node)):
        if node[i] is not None:
            if child < len(node):
                node[i].left = node[child]
                child += 1
            if child < len(node):
                node[i].right = node[child]
                child += 1
    return node[0]


def sum(root):
    def dfs(root, cur):
        if root is None:
            return 0

        cur = cur * 10 + root.val
        if root.left is None and root.right is None:
            return cur

        return dfs(root.left, cur) + dfs(root.right, cur)

    return dfs(root, 0)

r1 = inittree([1, 2, 3])
r2 = inittree([4, 9, 0, 5, 1])
print('输入：root = [1,2,3], 输出：' + str(sum(r1)))
print('输入：root = [4,9,0,5,1], 输出：' + str(sum(r2)))
````
